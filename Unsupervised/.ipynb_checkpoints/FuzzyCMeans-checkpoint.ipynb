# Fuzzy C Means

## Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import colorsys
import time
```

## Defining the algorithm

### Core idea
It is advised that you read through the K-Means clustering algorithm prior to this. Fuzzy C Means is an improved version of K-Means algorithm designed to combat the issue where a single centroid can hog up two or more clearly distinct clusters and reach an unfavourable local minimum with no escape.

It has an additional parameter to K-means - fuzziness ($m$)

The difference is in the way centroids are updated :

The position of centroid j, $\LARGE c_j = \frac{\sum_x w_j(x)^m \cdot x}{\sum_x w_j(x)^m}$

where, $\large w_j(x) = \frac{1}{\displaystyle\sum_{i = 1}^k \left(\frac{\left\|x - c_j\right\| }{\left\|x - c_i\right\| }\right)^\frac{2}{m-1}}$

### Why is Fuzzy C Means better than K-Means?

It is designed to solve the stranded centroid problem where even if you have the correct value for k, the initial setup might just be unlucky such that a centroid may be stranded while another centroid hogs up on two or more clusters for himself.

Consider this initial scenario. There are two true clusters 1 and 2. There are two centroids A and B. Centroid A is right in the middle of the clusters 1 and 2. Centroid B is far enough way such that 1 and 2 are both assigned to centroid A.

Now in K-Means, this would be the end. No new points will be assigned to B and we will exit the algorithm with a sub-optimal clustering.

But in Fuzzy C Means, B's position will be updated to be closer to the clusters even if it has no points its domain. B will keep edging to the clusters until both of them settle into their respective clusters. With enough time, this is almost guaranteed to happen unless the initial assignment is ridiculously unlucky.

```python
class FuzzyCMeans:
    def __init__(self, c, X, fuzziness, dims=2, seed=None):
        if seed is not None:
            np.random.seed(seed)
        self.m = fuzziness  # Fuzziness parameter. Used for determining the weights of each datapoint
        if self.m <= 1:  # Fuzziness parameter must be in the range (1, âˆž)
            self.m = 1.01  # Arbitrarily small m value
        self.X = X  # X is the dataset
        self.c = c  # c is the number of centroids
        self.n = X.shape[0]  # n is the number of points in the dataset
        self.dims = dims  # dims is the nmber of dimensiions of the dataset
        self.centroids = X[np.random.choice(self.n, self.c, replace=False)]  # c centroids randomly chosen from X
        self.prev_centroids = np.copy(
            self.centroids)  # Stores the previous centroids to determine when the centroids have converged
        self.y_pred = np.zeros((self.n, self.c),
                               dtype='float32') - 1.0  # Make an array of shape = (num of points, num of centroids) and fill it with -1

    # Assigns each point a list of weights for each centroid
    def assign(self):
        epsilon = 1e-9  # Small value to avoid division by zero error
        # Calculate distances between all points and centroids at once
        distances = np.linalg.norm(self.X[:, np.newaxis] - self.centroids, axis=2) + epsilon
        # Calculate the membership matrix
        self.y_pred = 1 / np.sum((distances[:, :, np.newaxis] / distances[:, np.newaxis, :]) ** (2 / (self.m - 1)),
                                 axis=2)
        return self.y_pred

    # Updates the location of the centroids based on the weights
    def update(self):
        self.prev_centroids = np.copy(self.centroids)
        for i in range(self.c):
            # Finds the new location of the centroids based on the weighted average of all the data points
            # The [: , np.newaxis] part is just to explicitly convert the Wij**m part to a coloumn vector. It works fine without it as well.
            self.centroids[i] = np.sum(self.X * (self.y_pred[:, i] ** self.m)[:, np.newaxis], axis=0) / np.sum(
                self.y_pred[:, i] ** self.m)

        return self.centroids

    # Runs the Fuzzy-C-Means algorithm
    def run(self, verbose=False):
        y_hist = [np.copy(self.y_pred), ]  # Keeps track of the weight list of the dataset
        centroid_hist = [np.copy(self.centroids), ]  # Keeps track of the centroid locations
        y_hist.append(np.copy(self.assign()))
        centroid_hist.append(np.copy(self.update()))

        # Alternates between assignment and updation
        i = 0
        while not np.allclose(self.prev_centroids, self.centroids):
            if i % 2 == 0:
                y_hist.append(np.copy(self.assign()))
            else:
                centroid_hist.append(np.copy(self.update()))
                # If verbose, it prints the current iteration
                if verbose:
                    print(f"Iteration {i // 2}")

            i += 1

        print("Fuzzy C-Means Converged.")
        return centroid_hist, y_hist
```

```python
WHITE = (255, 255, 255)
GREY = (195, 195, 195)
BLACK = (0, 0, 0)


# Color function to generate distinct colors based on a seed/number
# I stole this code. Don't know why or how it works. Just needed random colors
def colors(num):
    if num == -1:
        return GREY
    hue = (num * 137.508) % 360  # Use golden angle approximation and keep the hue within [0, 360)
    saturation = 0.8
    lightness = 0.5

    r, g, b = colorsys.hls_to_rgb(hue / 360, lightness, saturation)

    return np.array([r,g,b])

# Returns a gaussian mixture with desired number of clusters
def get_random_data(num_points, centers, seed=None):
    if seed is None:
        seed = np.random.randint(1, 1000)
    print(f"Seed : {seed}")
    X, Y = make_blobs(n_samples=num_points, n_features=2, centers=centers, cluster_std=1.0, random_state=seed)
    # Returns both the datapoints and the actual cluster numbers, although it is not used
    return X, Y, seed


# Displays the datapoints as well as the centroids (if any)
def show_data(X, Y_pred, centroids):
    plt.scatter(X[:,0],X[:,1],color = [colors(label) for label in Y_pred])
    plt.scatter(centroids[:,0], centroids[:,1], color = 'black', marker = 's', edgecolors = 'white', linewidths = 1)
    plt.show()


def setup_fuzzy_c_means(X, centers, m, seed=None, verbose=False):
    cluster = FuzzyCMeans(centers, X, m, dims=2, seed=seed)
    centroids_hist, y_pred_hist = cluster.run(verbose)
    y_pred_hist = np.argmax(y_pred_hist, axis=2)
    y_pred_hist[0] -= 1
    return centroids_hist, y_pred_hist
    

def distortion(X, y_pred, centroid):
    centroid = np.asarray(centroid)
    distortion = 0
    for i in range(X.shape[0]):
        # print(i, X[i], y_pred[i], centroid[y_pred[i]], np.linalg.norm(X[i] - centroid[y_pred[i]]))
        distortion += np.linalg.norm(X[i] - centroid[y_pred[i]])
    return distortion / X.shape[0]


def elbow_plot(X, m, k_low=2, k_high=2, plot=False):
    distortions = []
    for i in range(k_low, k_high + 1):
        centroids_hist, y_pred_hist = setup_fuzzy_c_means(X, i, m)
        distortions.append(distortion(X, y_pred_hist[-1], centroids_hist[-1], ))

    if plot:
        plt.plot(list(range(k_low, k_high + 1)), distortions)
        plt.xticks(list(range(k_low, k_high + 1)))
        plt.title('Elbow plot')
        plt.ylabel('Distortion')
        plt.xlabel('k')
        plt.show()

    return distortions


def menu_driven(default_params = None):
    if default_params is None:
        centers = int(input("Enter number of centers for the random data : "))
        
        fuzziness = float(input("Enter the fuzzines value : "))
    
        seed = input("Enter the seed if any. Type None for random seed : ")
        if seed.lower() == "none":
            seed = None
        else:
            seed = int(seed)
    
        num_points = int(input("Enter the number of points : "))
        
        elbow = input("Do you want an elbow plot? Y/N : ")
    else:
        centers, fuzziness, seed, num_points, elbow = default_params

    X, _, seed = get_random_data(num_points, centers, seed)
    if elbow.upper() == 'Y':
        low = int(input("Enter lower limit of centroids : "))
        high = int(input("Enter upper limit of centroids : "))
        elbow_plot(X, fuzziness, low, high, True)
        quit()
    else:
        centroids_hist, y_pred_hist = setup_fuzzy_c_means(X, centers, fuzziness)

    print(f"{len(y_pred_hist)} iterations to converge.")
    print(np.max(y_pred_hist) + 1, "clusters")

    return X, centroids_hist, y_pred_hist

op = input("Do you want default params? Y/N : ")
if op.lower() == 'y':
    def_params = [3, 3, None, 500, 'N']
else:
    def_params = None

X, centroids_hist, y_pred_hist = menu_driven(def_params)

show_data(X,np.asarray(y_pred_hist[-1]),np.asarray(centroids_hist[-1]))
```

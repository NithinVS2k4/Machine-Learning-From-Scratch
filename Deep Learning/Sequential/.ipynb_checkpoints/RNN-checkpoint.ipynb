#### Creating a future return column

```python
data_original = data
```

```python
# Contains the % change in closing value 5 days ahead.
# As a result, the last 5 values have to be omitted
closing_length = 5
data[f'Future_Return_{closing_length}'] = (data['Close'].shift(-closing_length) - data['Close']) / data['Close']

# Previous day close data will be used as a feature. The first sample has to be omitted as a result
data['Previous Close'] = data['Close'].shift(1)

data
```

```python
data = data.iloc[1:-closing_length,:].reset_index(drop=True)
data
```

```python
length_data = len(data)
split_ratio = 0.7

length_train = round(length_data * split_ratio)  
length_test = length_data - length_train

print("Data length :", length_data)
print("Train data length :", length_train)
print("Validation data lenth :", length_test)
```

```python
features = ['Open', 'Low', 'High', 'Close', 'Previous Close', 'Volume']
label = 'Future_Return_5'

for feature in features:
    data[feature] = (data[feature]-data[feature].min())/(data[feature].max() - data[feature].min())
```

```python

plt.style.use('Solarize_Light2')
plt.figure(figsize=(18, 6), dpi=80)
for feature in data.columns:   
    if feature in ['Date','Future_Return_5','Previous Close', 'Volume']: continue
    plt.plot(data['Date'].values, data[feature].values, label = feature)

plt.legend()
plt.show()
```

```python
X_train = data[:length_train][features]
X_train.head()
```

```python
X_test = data[length_train:][features]
X_test.head()
```

```python
y_train_full = data[:length_train][label].values
y_test_full = data[length_train:][label].values

y_test_full.shape
```

```python
X_train_rnn = []
y_train_rnn = []

time_interval = 30

for i in range(time_interval, length_train):
    X_train_rnn.append(X_train.iloc[i-time_interval:i,:])
    y_train_rnn.append(y_train_full[i])
    
X_train_rnn, y_train_rnn = np.array(X_train_rnn), np.array(y_train_rnn)
y_train_rnn = y_train_rnn.reshape(-1,1)
```

```python
X_test_rnn = []
y_test_rnn = []

time_interval = 30

for i in range(time_interval, length_test):
    X_test_rnn.append(X_test.iloc[i-time_interval:i,:])
    y_test_rnn.append(y_test_full[i])
    
X_test_rnn, y_test_rnn = np.array(X_test_rnn), np.array(y_test_rnn)
y_test_rnn = y_test_rnn.reshape(-1,1)
```

## Defining the RNN model

## BPTT Calculations

#### Loss function $\mathcal{L_t}$ = $-{\displaystyle \sum_i^{out\_size}} y_t^i \cdot ln \hat{y}_i^t$

#### where $\hat{y}_t^i = \frac{e^{o_t^i}}{\displaystyle \sum_j e^{o_t^j}}$

### Gradient Calculations

#### Gradient of $\hat{y}$
$\Huge
\frac{\partial L}{\partial \hat{y}}$

#### Gradient of $o_t$

First we calculate gradient of $o_t^i$ where $i \in [1,out\_size]$

$\Huge
\frac{\partial L}{\partial o_t^i} = -\frac{\partial}{\partial o_t^i} \displaystyle\sum_i y^i_t \cdot ln\hat{y}^i_t
$



#### Gradient of $V$

$\Huge
\frac{\partial L}{\partial V} = {\displaystyle\sum_{t=1}^{N}} \frac{\partial L}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial o_{t}} \cdot \frac{\partial o_{t}}{\partial V} 
$

$\Huge
\frac{\partial L}{\partial V} = {\displaystyle\sum_{t=1}^{N}} \frac{\partial L}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial o_{t}} \cdot h_t^T
$


#### Gradient of $W$
$\Huge 
\frac{\partial L}{\partial W} = \frac{\partial L}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial o_N} \cdot
\frac{\partial o_N}{\partial W} + \frac{\partial L}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial o_N} \cdot
\frac{\partial o_N}{\partial o_{N-1}} \cdot \frac{\partial o_{N-1}}{\partial W} + \frac{\partial L}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial o_N} \cdot \frac{\partial o_N}{\partial o_{N-1}} \cdot \frac{\partial o_{N-1}}{\partial o_{N-2}}
\cdot \frac{\partial o_{N-2}}{\partial W}  + ...
$

$\Huge
\frac{\partial L}{\partial U} = {\displaystyle \sum_{t=1}^{N}} \frac{\partial L}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial o_N} \cdot \frac{\partial o_{t}}{\partial W}
{\displaystyle \prod_{k=t}^{N-1}} \frac{\partial o_{k+1}}{\partial o_{k}}
$

#### Gradient of $h_t$

$\Huge
\frac{\partial L}{\partial h_t} = \frac{\partial L}{\partial o_t} \cdot \frac{\partial o_t}{\partial h_t} + \frac{\partial L}{\partial h_{t+1}} \cdot \frac{\partial h_{t+1}}{\partial h_t}
$

$\Huge
\frac{\partial L}{\partial h_t} = V^T\frac{\partial L}{\partial o_t} + W^T\frac{\partial L}{\partial h_{t+1}}
$

```python
import numpy as np

def softmax(x):
    p = np.exp(x - np.max(x))
    return p / np.sum(p)

class RNN:
    def __init__(self, hidden_size, input_size, output_size, act_func=np.tanh, task = "classification"):
        self.task = task
        
        self.hidden_size = hidden_size
        self.input_size = input_size
        self.output_size = output_size
        self.sigma = act_func  

        self.U = np.random.uniform(-1/np.sqrt(input_size), 1/np.sqrt(input_size), (hidden_size, input_size)) 
        self.W = np.random.uniform(-1/np.sqrt(hidden_size), 1/np.sqrt(hidden_size), (hidden_size, hidden_size))
        self.V = np.random.uniform(-1/np.sqrt(hidden_size), 1/np.sqrt(hidden_size), (output_size, hidden_size))

        self.b = np.zeros((hidden_size, 1))
        self.c = np.zeros((output_size, 1))
        
    def forward(self, x):
        N = len(x)
        self.h = np.zeros((N, self.hidden_size, 1)) 
        o = np.zeros((N, self.output_size, 1))  

        for t in range(N):
            xt = x[t].reshape(-1,1)
            self.h[t] = self.sigma(self.b + self.U @ xt + self.W @ self.h[t-1])  
            o[t] = self.c + self.V @ self.h[t]

        if self.task == "classification":
            return softmax(o[N-1])
        else:
            return o[N-1]

    def backward(self, x, y_hat, y, learning_rate=0.001):
        N = len(x)
        h = self.h

        dV = np.zeros_like(self.V)
        dW = np.zeros_like(self.W)
        dU = np.zeros_like(self.U)
        db = np.zeros_like(self.b)
        dc = np.zeros_like(self.c)

        dh_next = np.zeros((self.hidden_size, 1))

        for t in range(N):
            xt = x[t].reshape(-1,1)
            print(xt)
            break
```

```python
import numpy as np

# ---------------------------
# Regression Metrics Functions
# ---------------------------

def mse(y_true, y_pred):
    """Mean Squared Error"""
    return np.mean((y_true - y_pred) ** 2)

def rmse(y_true, y_pred):
    """Root Mean Squared Error"""
    return np.sqrt(mse(y_true, y_pred))

def mae(y_true, y_pred):
    """Mean Absolute Error"""
    return np.mean(np.abs(y_true - y_pred))

def r2_score(y_true, y_pred):
    """Coefficient of Determination"""
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - ss_res / ss_tot if ss_tot != 0 else 0

def directional_accuracy(y_true, y_pred):
    """
    Directional Accuracy: fraction of samples for which the sign of the prediction 
    matches the sign of the true value.
    """
    y_true = y_true.flatten()
    y_pred = y_pred.flatten()
    direction_true = np.sign(y_true)
    direction_pred = np.sign(y_pred)
    return np.mean(direction_true == direction_pred)

# ---------------------------
# Producing Predictions using the Trained RNN
# ---------------------------

# Assume X_test_rnn is a NumPy array of shape:
#   (num_samples, time_interval, input_size)
# And y_test_rnn is a NumPy array of shape (num_samples, 1)

# Example: Process each sequence (window) individually,
# and use the final time step's output as the prediction.
predictions = []

for i in range(len(X_test_rnn)):
    # Each sample (window) has shape (time_interval, input_size)
    # Use the forward method of your RNN model.
    y_hat_sequence = rnn.forward(X_test_rnn[i])
    # y_hat_sequence is an array of predictions per time step.
    # For a sequence-to-one model, take the final time step's prediction.
    prediction = y_hat_sequence[-1]
    predictions.append(prediction)

predictions = np.array(predictions).reshape(-1, 1)

# ---------------------------
# Evaluate the Model
# ---------------------------
mse_val = mse(y_test_rnn, predictions)
rmse_val = rmse(y_test_rnn, predictions)
mae_val = mae(y_test_rnn, predictions)
r2_val = r2_score(y_test_rnn, predictions)
dir_acc = directional_accuracy(y_test_rnn, predictions)

print("MSE:", mse_val)
print("RMSE:", rmse_val)
print("MAE:", mae_val)
print("R^2:", r2_val)
print("Directional Accuracy:", dir_acc)
```
